```python
from nltk.tokenize import sent_tokenize
```

```python
sent_tokenize(teste)
```

```python
tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')
```

```python
from nltk.tokenize import regexp_tokenize
```

```python
teste_tokenizado = regexp_tokenize(teste, "[\w']+")
```

```python
from nltk.corpus import stopwords
```

```python
english_stops = set(stopwords.words('english'))
```

```python
teste_finalizado = [teste_tokenizado for teste_tokenizado in teste_tokenizado if teste_tokenizado not in english_stops]
```

```python
teste_finalizado
```

Lematização

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```python
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
teste_lematizado = []

for palavra in teste_finalizado:
    teste_lematizado.append(wordnet_lemmatizer.lemmatize(palavra))
```

```python
teste_lematizado
```

Stemming

```python
from nltk.stem import RegexpStemmer
```

```python
regexp_stemmer = RegexpStemmer('ing')
```

```python
from nltk.stem import PorterStemmer
```

```python
porter_stemmer = PorterStemmer()
```

```python
teste_stemming = []
for palavra in teste_lematizado:
    teste_stemming.append(porter_stemmer.stem(palavra.lower()))
```

```python
teste_stemming
```

```python
Id = []
textos_tratados = []

for linha in range(len(df[0:])):
    texto = df.iloc[linha]['Text']
    texto_tokenizado = regexp_tokenize(str(texto), "[\w']+")
    texto_stop = [texto_tokenizado for texto_tokenizado in texto_tokenizado if texto_tokenizado not in english_stops]
    
    texto_lematizado = []
    for palavra in texto_stop:
        texto_lematizado.append(wordnet_lemmatizer.lemmatize(str(palavra).lower()))
    
    texto_stemming = []
    for palavra in texto_lematizado:
        texto_stemming.append(porter_stemmer.stem(str(palavra).lower()))
        
    Id.append(df.iloc[linha]['ID'])
    textos_tratados.append(texto_stemming)
    
```
